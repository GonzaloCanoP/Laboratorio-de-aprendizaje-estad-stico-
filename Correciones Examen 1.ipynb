```python
# Cargar datos
import pandas as pd

datos = pd.read_csv('imdb_top1000_lae.csv')
datos
```

```python
datos = datos.dropna()
datos = datos.drop_duplicates()
datos
```

### Quitar columnas

Decidi quitar las columnas de unnames porque no es relevante, certificate porque considero que no es relevante para predecir la calificación y solo puede complicar el modelo, series title y director porque si las convierto a dummies quedarian demasiadas variables y solo complicarian al modelo, y gross no considero que es relevante porque queremos predecir la calificación no el exito de la pelicula en taquilla

```python
datos = datos.drop(columns=['Unnamed: 0','Certificate','Series_Title','Director', 'Gross'])
datos.describe()
```

### Transformar datos


```python
# Released_Year -> quitar "PG" y convertir a entero
datos = datos[datos['Released_Year'] != "PG"]
datos['Released_Year'] = datos['Released_Year'].astype(int)

# Runtime -> aseguramos que sea string antes de reemplazar
datos['Runtime'] = datos['Runtime'].astype(str).str.replace(' min', '', regex=False)
datos['Runtime'] = datos['Runtime'].astype(int)
```

```python
datos.describe()
```

```python
datos.info()
```

```python
import numpy as np
from sklearn.metrics import r2_score
import statsmodels.api as sm
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.preprocessing import StandardScaler
```

```python
y = datos['IMDB_Rating']
X = datos.drop(columns=['IMDB_Rating'])
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size=0.6, random_state=137)
```

```python
# Escalamiento
scaler = StandardScaler().fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled =scaler.transform(X_test)
```

```python
# Regresión con Datos train
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)

n, p = X_train_scaled.shape

# Predicciones
y_hat = lr.predict(X_train_scaled)

# R2 (train)
R2 = r2_score(y_train, y_hat)


# RSS (train)
RSS = np.sum((y_train - y_hat)**2)

# RSE con gl = n - p - 1
RSE = np.sqrt(RSS / (n - p - 1))

# Var_beta
X = np.column_stack([np.ones(n), X_train_scaled])  # X con intercepto (train)

var_beta = np.linalg.pinv(X.T @ X) * (RSE**2)

std_beta = np.sqrt(var_beta.diagonal())

# Estadísticos t
beta = np.r_[lr.intercept_, lr.coef_]   # vector (p+1,)
t_values = beta / std_beta

# P-values (bilaterales) con gl = n - p - 1
p_values = 2 * (1 - stats.t.cdf(np.abs(t_values), df=n - p - 1))

print('---Resultados Regresión sin penalización con datos de entrenamiento---')
print(f"Intercepto: {lr.intercept_}")
print(f"Coeficientes: {lr.coef_}")
print(f"R2: {R2}")
print(f"P-values: {p_values}")
```

```python
# Con datos de prueba
n, p = X_test_scaled.shape

# Predicciones
y_hat = lr.predict(X_test_scaled)

# R2 (train)
R2 = r2_score(y_test, y_hat)


# RSS (train)
RSS = np.sum((y_test - y_hat)**2)

# RSE con gl = n - p - 1
RSE = np.sqrt(RSS / (n - p - 1))

# Var_beta
X = np.column_stack([np.ones(n), X_test_scaled])

var_beta = np.linalg.pinv(X.T @ X) * (RSE**2)

std_beta = np.sqrt(var_beta.diagonal())

# Estadísticos t
beta = np.r_[lr.intercept_, lr.coef_]   # vector (p+1,)
t_values = beta / std_beta

# P-values (bilaterales) con gl = n - p - 1
p_values = 2 * (1 - stats.t.cdf(np.abs(t_values), df=n - p - 1))

print('---Resultados Regresión sin penalización con datos de prueba---')
print(f"Intercepto: {lr.intercept_}")
print(f"Coeficientes: {lr.coef_}")
print(f"R2: {R2}")
print(f"P-values: {p_values}")
```

```python
# Regresión Lasso para datos de entrenamiento y prueba
L1 = Lasso(alpha=0.0001)
L1.fit(X_train_scaled, y_train)

#Entrenamiento
y_hat_train = L1.predict(X_train_scaled)
R2_train = r2_score(y_train, y_hat_train)

#Prueba
y_hat_test = L1.predict(X_test_scaled)
R2_test = r2_score(y_test, y_hat_test)

print('---Resultados Regresión Lasso con datos de entrenamiento y prueba---')
print(f"Intercepto: {L1.intercept_}")
print(f"Coeficientes: {L1.coef_}")
print(f"R2 train: {R2_train}")
print(f"R2 test: {R2_test}")
```

```python
# Regresión Ridge para datos de entrenamiento
L2 = Ridge(alpha=0.0001)
L2.fit(X_train_scaled, y_train)

y_hat_train = L2.predict(X_train_scaled)
R2_train = r2_score(y_train, y_hat_train)

y_hat_test = L2.predict(X_test_scaled)
R2_test = r2_score(y_test, y_hat_test)

print('---Resultados Regresión Lasso con datos de entrenamiento y prueba---')
print(f"Intercepto: {L2.intercept_}")
print(f"Coeficientes: {L2.coef_}")
print(f"R2 train: {R2_train}")
print(f"R2 test: {R2_test}")
```

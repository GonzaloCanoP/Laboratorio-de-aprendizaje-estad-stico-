## 1.1 Regresión con MPG como salida

```python
import pandas as pd
datos = pd.read_excel('Motor Trend Car Road Tests.xlsx')
datos = datos.drop(columns = 'model')
datos
```

```python
# Primera regresión

import statsmodels.api as sm
import numpy as np

y = datos['mpg']
x = datos.drop(columns = 'mpg')
n = len(datos)
unos = np.ones([n,1])

X=np.hstack([unos,x])
ols = sm.OLS(y,X)
results = ols.fit()
results.summary()
```

**Interpretación resultados regresión **

* R^2: podemos ver que tenemos una R^2 de 0.869 la cual nos indica que tiene un buen ajuste pero viendo los valores de los coeficientes podemos ver que la gran mayoria no son estadisticamente significativos, salvo x5 que es el que tiene el p-value más bajo y pudiera ser significativo si bajara ese p-value
* También podemos ver que la mayoria tiene signos negativos o son menores a 1 lo que nos indica que afectan de manera negativa


## Train test split

```python
# Train

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(x, y, train_size=0.4,random_state=137)

# Esclamiento (preguntar al profe)
scaler = StandardScaler().fit(X_train)
X_scaled_train= scaler.transform(X_train)

n=len(X_scaled_train)
unos = np.ones([n,1])

X = np.hstack([unos, X_scaled_train])
ols = sm.OLS(y_train,X)
results = ols.fit()
results.summary()
```

**Interpretación datos train**

* Con los datos train podemos ver que tenemos un R^2 casi perfecto de 0.99 el cual nos indica que el modelo es casi perfecto o puede estar sobreajustado, pero a pesar de eso si vemos los valores de los coeficientes tenemos algunos valores muy elevados y algunos valores negativos o menores a 0 y sus p-values nos muestran que no son estadisticamente significativos

```python
# Test
scaler = StandardScaler().fit(X_test)
X_scaled_test= scaler.transform(X_test)

n=len(X_scaled_test)
unos = np.ones([n,1])

X = np.hstack([unos, X_scaled_test])
ols = sm.OLS(y_test,X)
results = ols.fit()
results.summary()
```

**Interpretación datos test**

* La R^2 de los datos test sigue siendo muy alta y nos sigue mostrando que el modelo tiene un buen ajuste pero si vemos los p-values de las variables podemos ver que solo la constante es estadisticamente significativa y los coeficientes no.
* La mayoria de los signos son negativos y los valores menores a 1 lo cual nos muestra que la mayoria de los coeficientes nos afectan de manera negativa al resultado

```python
# Regresión con penalización L2
from sklearn.linear_model import Ridge
from scipy import stats

y = datos['mpg']
x = datos.drop(columns = 'mpg')
n,p = x.shape

L2 = Ridge(alpha=0.01, fit_intercept=True)
L2.fit(x,y)

# RSS
y_hat = L2.predict(x)
RSS_L2 = np.sum((y-y_hat)**2)

#RSE
RSE_L2 = np.sqrt(RSS_L2/(n-p))

# Var_beta
X = np.column_stack([np.ones(n),x]) # Es la x con el intercepto
var_beta = np.linalg.inv(X.T @ X) * RSE_L2**2

std_beta = np.sqrt(var_beta.diagonal())

# Estadisticos t
beta = np.r_[L2.intercept_, L2.coef_]   # vector (p+1,)
t_L2 = beta / std_beta

# P-values
p_L2 = 2 * (1 - stats.t.cdf(np.abs(t_L2), n-p-1))

# R^2
R2 = L2.score(x, y)

print("----Resultados Ridge(L2)----")
print(f"Intercepto: {L2.intercept_}")
print(f"Coefficientes: {L2.coef_}")
print(f"RSS: {RSS_L2}")
print(f"RSE: {RSE_L2}")
print(f"P-Values: {p_L2}")
print(f"R^2: {R2}")
```

**Interpretación regresión con penalización**

* Podemos ver que en este caso la R^2 bajo pero los valores de los p-values nos siguen mostrando que los datos no son estadisticamente significativos y seguimos teniendo valores negativos y menores a 1 en los coeficientes que no aportan nada positivo a los resultados.

```python
y = datos['qsec']
x = datos.drop(columns = 'qsec')
n,p = x.shape

L2 = Ridge(alpha=0.05, fit_intercept=True)
L2.fit(x,y)

# RSS
y_hat = L2.predict(x)
RSS_L2 = np.sum((y-y_hat)**2)

#RSE
RSE_L2 = np.sqrt(RSS_L2/(n-p))

# Var_beta
X = np.column_stack([np.ones(n),x]) # Es la x con el intercepto
var_beta = np.linalg.inv(X.T @ X) * RSE_L2**2

std_beta = np.sqrt(var_beta.diagonal())

# Estadisticos t
beta = np.r_[L2.intercept_, L2.coef_]   # vector (p+1,)
t_L2 = beta / std_beta

# P-values
p_L2 = 2 * (1 - stats.t.cdf(np.abs(t_L2), n-p-1))

# R^2
R2 = L2.score(x, y)

print("----Resultados Ridge(L2)----")
print(f"Intercepto: {L2.intercept_}")
print(f"Coefficientes: {L2.coef_}")
print(f"RSS: {RSS_L2}")
print(f"RSE: {RSE_L2}")
print(f"P-Values: {p_L2}")
print(f"R^2: {R2}")
```

**Interpretación con penalización**

* También en este caso tenemos valores muy parecidos que los datos de test y la normal, lo cual nos muestra que podriamos usar cualquiera de esos tres metodos, menos train ya que nos muestra un sobreajuste

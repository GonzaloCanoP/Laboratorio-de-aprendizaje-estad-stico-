```python
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
data = pd.read_csv('Advertising.csv', index_col=0)
data
```

```python
data.info()
```

---

## Primera regresion simple

```python
# Primera regresión
y = data['sales']
x = data.drop(columns=['sales'])
lr = LinearRegression()
lr.fit(x, y)

y_pred = lr.predict(x)
r2 = r2_score(y, y_pred)
print(r2)
```

```python
# Decision Tree
arbol = DecisionTreeRegressor()
arbol.fit(x, y)
y_pred = arbol.predict(x)
r2 = r2_score(y, y_pred)
print(r2)
```

---

## Con train y test

```python
# Regresión con train test split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size = 0.6, random_state = 42)
lr = LinearRegression()
lr.fit(x_train, y_train)

#Test
y_pred_test = lr.predict(x_test)
r2_test = r2_score(y_test, y_pred_test)

# Train
y_pred_train = lr.predict(x_train)
r2_train = r2_score(y_train, y_pred_train)

print(f"R^2 Test: {r2_test}")
print(f"R^2 Train: {r2_train}")
```

```python
# Arbol con train y test
arbol = DecisionTreeRegressor()
arbol.fit(x_train, y_train)

# Test
y_pred_test = arbol.predict(x_test)
r2_test = r2_score(y_test, y_pred_test)
print(f"R^2 Test: {r2_test}")

# Train
y_pred_train = arbol.predict(x_train)
r2_train = r2_score(y_train, y_pred_train)
print(f"R^2 Train: {r2_train}")
```

---

## Con PolynomialFeatures

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
# Con todos los datos
poly_lr = Pipeline([
    ("scale", StandardScaler()),
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("lin", LinearRegression())
])

poly_lr.fit(x, y)
r2_poly = poly_lr.score(x, y)
print(f"R^2 polinomial: {r2_poly}")
```

```python
# Arbol con todos los datos
arbol_poly = Pipeline([
    ("scale", StandardScaler()),
    ("poly", PolynomialFeatures(degree=2, include_bias=False)),
    ("tree", DecisionTreeRegressor())
])
arbol_poly.fit(x, y)
r2_poly = arbol_poly.score(x, y)
print(f"R^2 polinomial del arbol: {r2_poly}")
```

```python
# Regresion polinomial con datos train y test
poly_lr.fit(x_train, y_train)
r2_poly_train = poly_lr.score(x_train, y_train)
print(f"R^2 polinomial (train): {r2_poly_train}")

r2_poly_test = poly_lr.score(x_test, y_test)
print(f"R^2 polinomial (test): {r2_poly_test}")
```

```python
# Arbol con train y test
arbol_poly.fit(x_train, y_train)
r2_poly_train = arbol_poly.score(x_train, y_train)
print(f"R^2 arbol polinomial (train): {r2_poly_train}")
r2_poly_test = arbol_poly.score(x_test, y_test)
print(f"R^2 arbol polinomial (test): {r2_poly_test}")
```
